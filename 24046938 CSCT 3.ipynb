```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyArrow

# to create figure
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.axis("off")

# to draw boxes
def draw_box(x, y, text, width=2.5, height=1.0):
    rect = Rectangle((x, y), width, height, linewidth=2, edgecolor="black", facecolor="#cce5ff")
    ax.add_patch(rect)
    ax.text(x + width/2, y + height/2, text, ha="center", va="center", fontsize=10, wrap=True)

# to draw arrows
def draw_arrow(x1, y1, x2, y2):
    ax.add_patch(FancyArrow(x1, y1, x2-x1, y2-y1, 
                            width=0.02, head_width=0.3, head_length=0.3, 
                            length_includes_head=True, color="black"))

# to draw the components
draw_box(1, 6, "Data Sources\n(Environment Agency)")
draw_box(4, 6, "Preprocessing\n(Cleaning & Feature Eng.)")
draw_box(7, 6, "Flood Prediction Model\n(Random Forest)")

draw_box(2.5, 3.5, "Streamlit Application\n(User Interface)")

draw_box(6, 3.5, "Early Warning System\n(Alerts & Logging)")

draw_box(4, 1, "Outputs\n(Maps, Predictions, CSV Logs)")

# Draw arrows
draw_arrow(3.5, 6.5, 4, 6.5)  # data -> preprocessing
draw_arrow(6.5, 6.5, 7, 6.5)  # preprocessing -> model
draw_arrow(5.25, 6, 3.75, 4.5) # preprocessing -> app
draw_arrow(7.75, 6, 7, 4.5)    # model -> EWS
draw_arrow(5, 3.5, 5, 2)       # app -> outputs
draw_arrow(7.5, 3.5, 5.5, 2)   # EWS -> outputs

plt.tight_layout()
plt.savefig("flood_system_architecture_matplotlib.png", dpi=300)
plt.title("System Architecture Diagram", fontsize=16, weight="bold")
plt.show()
```

```python
from matplotlib.patches import FancyBboxPatch

# function to draw larger boxes with readable text
def draw_box(ax, text, center, boxstyle="round,pad=0.5", fontsize=14, width=0.4, height=0.1):
    x, y = center
    box = FancyBboxPatch(
        (x - width/2, y - height/2), width, height,
        boxstyle=boxstyle, linewidth=1.5, edgecolor="black", facecolor="#cce5ff"
    )
    ax.add_patch(box)
    ax.text(x, y, text, ha="center", va="center", fontsize=fontsize, weight="bold")

# Creating figure
fig, ax = plt.subplots(figsize=(6, 10))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

steps = [
    "Data Sources\n(Environment Agency)",
    "Data Preprocessing\n(Cleaning & Alignment)",
    "Model Training\n(Random Forest)",
    "Flood Prediction Pipeline",
    "Streamlit App\n(User Interface)",
    "Outputs\n(Maps, Alerts, CSV Logs, EWS)"
]

# for the boxes
y_positions = list(reversed([i/6 + 0.1 for i in range(len(steps))]))
for step, y in zip(steps, y_positions):
    draw_box(ax, step, (0.5, y), fontsize=14, width=0.8, height=0.1)

# for drawing arrows
for i in range(len(steps)-1):
    ax.annotate(
        "", xy=(0.5, y_positions[i+1]+0.05), xytext=(0.5, y_positions[i]-0.05),
        arrowprops=dict(arrowstyle="->", lw=1.5, color="black")
    )

plt.title("Flood Prediction System Flow", fontsize=16, weight="bold")
plt.show()
```
